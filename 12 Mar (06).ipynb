```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set the degrees of freedom
df = 10

# Generate x values for the plot
x = np.linspace(0, 30, 500)

# Generate the chi-square distribution with 10 degrees of freedom
chi_squared = stats.chi2.pdf(x, df)

# Create the plot
fig, ax = plt.subplots(figsize=(8,6))

# Plot the chi-square distribution
ax.plot(x, chi_squared, label="Chi-square Distribution (df={})".format(df))

# Shade the area corresponding to a chi-square statistics of 15
shade = np.linspace(15, 30, 100)
ax.fill_between(shade, stats.chi2.pdf(shade, df), color='blue', alpha=0.5)

# Set the axis labels and title
ax.set_xlabel("Chi-square statistics")
ax.set_ylabel("Probability density")
ax.set_title("Chi-square Distribution (df={})".format(df))

plt.legend()
plt.show()
```
